```python
#Cleaning the Data
```

```python
#First we see that x,y and z coloumns we had been used them with the new attributes so we need to drop them
#Also the coloumn untitles we must delete it
```

```python
df.drop(['x','y','z','Unnamed: 0'],axis=1,inplace=True)
df.head()
```

```python
#Check if there is NAN values
df.isna().sum()
```

```python
#No NAN values
```

```python
#Check if there are zeros 
(df==0).sum(axis=0)
```

```python
#We have zeroes so we could delete it or replace it with mean/median
```

```python
m1=df['Area'].median()
m2=df['Volume'].median()

k1=(df['Area']==0)
k2=(df['Volume']==0)
df[k1]=m1
df[k2]=m2
```

```python
(df==0).sum(axis=0)
#No Zeroes values
```

```python
#Dealing with categories attributes
```

```python
#Transfer the categories into Ordinary Encoding
from sklearn.preprocessing import OrdinalEncoder
cat1=df['cut']
cat2=df['color']
cat3=df['clarity']
cat1=pd.get_dummies(cat1)
cat2=pd.get_dummies(cat2)
cat3=pd.get_dummies(cat3)
OE=OrdinalEncoder()
oe_cut=OE.fit_transform(cat1)
oe_color=OE.fit_transform(cat2)
oe_clarity=OE.fit_transform(cat3)
#We dont need to transform it to array
```

```python
#Dealing with outliers
```

```python
#Finding IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR=Q3-Q1
```

```python
#Detecting outliers and erasing them
df_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df_out.shape
```

```python
#Split Data into Train & Test sets 
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df,test_size=0.2,random_state=42)
```

```python
df_labels=df["price"].copy()
df=train_set.drop("price",axis=1)
df_num = df.drop(["cut", "color", "clarity"], axis = 1)
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
df_scaled=std.fit_transform(df_num)
```

```python
#Display scaled Data
pd.DataFrame(df_scaled).head()
```

```python
#Transformation pipeline
from sklearn.compose import ColumnTransformer
num_attribs = list(df_num)
cat_attribs = ["cut","color","clarity"]
full_pipeline = ColumnTransformer([
 ("num", std, num_attribs),
 ("cat", OrdinalEncoder(), cat_attribs),
 ])
df_prepared = full_pipeline.fit_transform(df_out)
```

```python
df_prepared
```

```python
df_labels=df_out["price"].copy()
```

```python
#Select and train model
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(df_prepared, df_labels)
```

```python
from sklearn.metrics import mean_squared_error
df_predictions = tree_reg.predict(df_prepared)
tree_mse = mean_squared_error(df_labels, df_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
#Try cross validation
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, df_prepared, df_labels,
 scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```
